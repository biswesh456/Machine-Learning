```python
import gym
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
env = gym.make('FrozenLake-v0')
```

```python
class Q_model(nn.Module):
  def __init__(self, input_size, output_size):
    super().__init__()
    self.in_size = input_size
    self.out_size = output_size
    self.fc = nn.Linear(self.in_size, self.out_size)
    
  def forward(self, x):
#     x = torch.tensor(x, dtype=torch.float)
    x = self.fc(x)
    return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
Model = Q_model(16, 4).to(device)
optimizer = torch.optim.Adam(Model.parameters(), lr=0.1)
criterion = nn.MSELoss()
```

```python
def one_hot(x, input_size):
  lst = torch.zeros(input_size)
  lst[x] = 1
  return lst
```

```python
gamma=0.99
e=0.1
num_episodes=4000
jList = []
rList = []
for i in range(num_episodes):
  s=env.reset()
  rAll=0
  d=False
  j=0
  while j<99:
    j+=1
    # convert the states to one hot vectors
    s_vector = one_hot(s, 16).to(device)
    Q_for_s = Model(s_vector)
    # get the argmax of the Q i.e the action maximizing the Q of s
    _, a = Q_for_s.max(0)
    a = a.item()
    
    # to explore we occasionally take random a
    if np.random.rand(1) < e:
      a = env.action_space.sample()
      
    next_s,r,d,_ = env.step(a)
    
    next_s_vector = one_hot(next_s, 16).to(device)
    # get the max value of Q for next_s
    next_Q_maxvalue, _ = Model(next_s_vector).max(0)
    
    target_Q_for_s = Q_for_s
    target_Q_for_s[a] = r + gamma*next_Q_maxvalue
    loss = criterion(Q_for_s, target_Q_for_s)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    rAll += r
    s = next_s
    if d == True:
        #Reduce chance of random action as we train the model.
        e = 1./((i/50) + 10)
        break
  jList.append(j)
  rList.append(rAll)

print("Percent of succesful episodes: " + str(sum(rList)/num_episodes) + "%") 
```

```python
plt.plot(rList)
```

**Without neural network with just table implementation**

```python
total_states = env.observation_space.n
total_actions = env.action_space.n
Q = np.zeros([total_states, total_actions])
```

```python
plt.plot(jList)
```

```python
lr=0.8
gamma=0.95
num_episodes=2000
```

```python
reward_list = []
for i in range(num_episodes):
  s=env.reset()
  rAll=0
  d=False
  j=0
  while j<99:
    j+=1
    env.render()
    #Choose an action by greedily (with noise decreasing with episodes) 
    #picking from Q table
    a = np.argmax(Q[s,:] + np.random.randn(1,env.action_space.n)*(1./(i+1)))
    #Get new state and reward from environment
    new_s,r,d,_ = env.step(a)
    #Update Q-Table with new knowledge
    Q[s,a] = Q[s,a] + lr*(r + gamma*np.max(Q[new_s:])-Q[s,a])
    rAll+=r
    s=new_s
    
    if d == True:
      break
  
  reward_list.append(rAll)
    
```
